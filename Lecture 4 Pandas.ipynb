## Arithmetic Operations

```python
import pandas as pd
import numpy as np
```

```python
s1 = pd.Series(np.arange(4), index = ["a","c","d","e"])
s2 = pd.Series(np.arange(5), index = ["a","c","e","f","g"])
print(s1)
print(s2)
```

```python
# s1 + s2
s1 + s2
```

```python
df1 = pd.DataFrame(
    np.arange(6).reshape(2,3),
    columns = list("ABC"),
    index = ["Tim","Tom"])
df2 = pd.DataFrame(
    np.arange(9).reshape(3,3),
    columns = list("ACD"),
    index = ["Tim","Kate","Tom"])
print(df1)
print(df2)
```

```python
# Plus df1 with df2
df1 + df2
```

```python
df1.add(df2,fill_value = 0)
```

```python
# devide 1 -> df1
1/df1
```

```python
# df1 multiple by 3
df1 *3
```

```python
s = df2.iloc[1]
s
```

```python
print(df2)
df2 - s
```

## Applying a Function (Thiết kế a lambda hoặc function và apply)
